```python
import os
import shutil
import json

# 定义根目录
root_dir = "/data/yisi/mywork/SV3D-fine-tune/input_data/04ed/"

# 遍历根目录下的文件
for filename in os.listdir(root_dir):
    if filename.endswith(".png"):
        # 获取文件的编号 (000, 001, ...)
        file_prefix = filename.split('.')[0]
        
        # 创建对应的文件夹
        new_dir = os.path.join(root_dir, file_prefix)
        os.makedirs(new_dir, exist_ok=True)
        
        # 移动文件
        for ext in [".png", ".npy", ".pt"]:
            old_path = os.path.join(root_dir, f"{file_prefix}{ext}")
            new_path = os.path.join(new_dir, f"{file_prefix}{ext}")
            if os.path.exists(old_path):
                shutil.move(old_path, new_path)

print("文件结构调整完成。")
```

```python
import os
import shutil

# 定义根目录
root_dir = "/data/yisi/mywork/SV3D-fine-tune/input_data/04ed/"

# 遍历根目录下的子文件夹
for subdir in os.listdir(root_dir):
    subdir_path = os.path.join(root_dir, subdir)
    
    if os.path.isdir(subdir_path):
        # 移动子文件夹中的所有文件回根目录
        for filename in os.listdir(subdir_path):
            old_path = os.path.join(subdir_path, filename)
            new_path = os.path.join(root_dir, filename)
            
            # 移动文件
            if os.path.exists(old_path):
                shutil.move(old_path, new_path)
                print(f"Moved: {old_path} to {new_path}")
        
        # 删除子文件夹
        os.rmdir(subdir_path)
        print(f"Deleted directory: {subdir_path}")
        
        # 删除cardiac.json文件
        cardiac_json_path = os.path.join(subdir_path, 'cardiac.json')
        if os.path.exists(cardiac_json_path):
            os.remove(cardiac_json_path)
            print(f"Deleted cardiac.json: {cardiac_json_path}")

print("文件恢复到调整前的状态完成。")
```
